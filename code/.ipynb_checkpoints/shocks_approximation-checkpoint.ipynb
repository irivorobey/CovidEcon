# Построение коэффициентов локализации

```python
def get_t(labour, idx_region, delta):
    t_ij = np.zeros((14, 14))
    for i in range(0, 14):
        for j in range(0, 14):
            if i == j:
                t_ij[i, j] = (labour[idx_region, i] / labour[idx_region, :].sum() *
                              labour[0, :].sum() / labour[0, i])
            else:
                t_ij[i, j] = (labour[idx_region, i] / labour[0, i] *
                              labour[0, j] / labour[idx_region, j])
    lam = np.log2(1 + labour[idx_region, :].sum() / labour[0, :].sum()) ** delta
    t_ij = np.minimum(t_ij * lam, np.ones((14, 14)))
    return t_ij
```

```python
def get_labour(year, df_labour_region, df_labour_branch):
    branch = np.array(df_labour_branch)[:, 1:]
#     print(branch)
    region = np.tile(df_labour_region.loc[:, year], (14, 1)).T
#     print(region)
    return branch * region / 100
```

```python
T = []
L = []
for i in np.arange(2015, 2021):
    idx = i - 2015
    labour = get_labour(i, df_labour_region, labour_branch[idx])
    L.append(labour)
    T.append(get_t(labour, 76, 0.12))
```

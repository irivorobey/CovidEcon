```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('MOB_Novosib.xlsx', sheet_name=1, index_col=0)
df
```

```python
mob = np.array(df)
mob = mob[:-1, :]
```

```python
G = nx.from_numpy_array(mob, create_using=nx.DiGraph)
layout = nx.circular_layout(G)
nx.draw(G, layout)
# nx.draw_networkx_edge_labels(G, pos=layout)
plt.show()
```

```python
mob.sum(0)
```

```python
(mob.sum(0) - mob.sum(1)) / mob.sum(0)
```
